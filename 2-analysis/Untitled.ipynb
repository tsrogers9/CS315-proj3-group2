```python
import pandas as pd

# Replace 'file_path.csv' with your file's path
male_all_description = pd.read_csv('M_all_vid_descriptions_2024_04_28.csv')
```

```python
male_all_description
```

```python
male_categories = pd.read_csv('male_categories_may2.csv')
```

```python
male_categories
```

```python
male_categories.rename(columns={'Description': 'Category'}, inplace=True)
```

```python
male_categories
```

```python
mask = ~male_categories['video_id'].isin(male_all_description['video_id'])

# Filter male_categories with the mask
non_matching_categories = male_categories[mask]
```

```python
non_matching_categories
```

```python
male_categories_filtered = male_categories[~male_categories['video_id'].isin(non_matching_categories['video_id'].tolist())]
```

```python
male_categories_filtered
```

```python
merged_male = pd.merge(male_categories_filtered, male_all_description, on='video_id', how='outer')
```

```python
merged_male 
```

```python
males_with_na = merged_male[merged_male['Category'].isna()]
```
